**9.1.Step-by-Step Code to Clean Titanic Dataset**

```python
df.info()
```

```python
print("\nMissing Values (Before Cleaning):")
print(df.isnull().sum())
```

```python
#Drop irrelevant columns (optional based on experiment goal)
df = df.drop(columns=['deck', 'embark_town', 'alive', 'class', 'who', 'adult_male', 'alone'])
```

```python
#Fill missing values
# Fill age with mean
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
# Fill embarked with mode
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
```

```python
# Drop rows with any remaining missing values
df = df.dropna()
```

```python
# Convert categorical columns into numeric using encoding (optional)
df['sex'] = df['sex'].map({'male': 0, 'female': 1})
df['embarked'] = df['embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
# View cleaned data info
print("\nCleaned Dataset Info:")
print(df.info())
```

```python
# View missing values after cleaning
print("\nMissing Values (After Cleaning):")
print(df.isnull().sum())
```

```python
# Save cleaned dataset to CSV and Excel
df.to_csv("cleaned_titanic.csv", index=False)
df.to_excel("cleaned_titanic.xlsx", index=False)
```

```python
# Step 1 Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2 Load the Titanic Dataset
df = sns.load_dataset('titanic')

print("Dataset Head")
print(df.head())

print("\nDataset Tail")
print(df.tail())

# Step 3 Check Dataset Shape
print("\nDataset Shape")
print(df.shape)

# Step 4 Summary Statistics
print("\nSummary Statistics")
print(df.describe(include='all'))

# Step 5 Dataset Information
print("\nDataset Information")
df.info()

# Step 6 Check Missing Values
print("\nMissing Values Before Cleaning")
print(df.isnull().sum())

# Step 7 Drop Irrelevant Columns
df = df.drop(columns=['deck','embark_town','alive','class','who','adult_male','alone'])

# Step 8 Handle Missing Values
df['age'] = df['age'].fillna(df['age'].mean())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

df = df.dropna()

# Step 9 Encode Categorical Columns
df['sex'] = df['sex'].map({'male':0,'female':1})
df['embarked'] = df['embarked'].map({'S':0,'C':1,'Q':2})

# Step 10 Verify Cleaned Dataset
print("\nCleaned Dataset Info")
df.info()

print("\nMissing Values After Cleaning")
print(df.isnull().sum())

# Step 11 Save Cleaned Dataset
df.to_csv("cleaned_titanic.csv", index=False)
df.to_excel("cleaned_titanic.xlsx", index=False)

# Step 12 Data Visualization
plt.figure(figsize=(15,10))

plt.subplot(2,3,1)
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival Count by Gender")

plt.subplot(2,3,2)
df['age'].hist(bins=20, edgecolor='black')
plt.title("Age Distribution")

plt.subplot(2,3,3)
sns.histplot(df['fare'], kde=True)
plt.title("Fare Distribution")

plt.subplot(2,3,4)
sns.boxplot(x='pclass', y='age', data=df)
plt.title("Age by Passenger Class")

plt.subplot(2,3,5)
sns.scatterplot(x='age', y='fare', hue='survived', data=df)
plt.title("Age vs Fare")

plt.tight_layout()
plt.show()

# Step 13 Additional Visualizations

sns.barplot(data=df, x='pclass', y='survived')
plt.title("Survival Rate by Passenger Class")
plt.show()

plt.pie(df['survived'].value_counts(),
        labels=['Not Survived','Survived'],
        autopct='%1.1f%%')
plt.title("Overall Survival Distribution")
plt.show()

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='magma')
plt.title("Correlation Heatmap")
plt.show()

sns.scatterplot(data=df, x='age', y='fare', hue='pclass', style='survived')
plt.title("Age vs Fare by Class")
plt.show()

sns.swarmplot(data=df, x='pclass', y='age', hue='sex')
plt.title("Age Distribution by Class and Gender")
plt.show()
```
